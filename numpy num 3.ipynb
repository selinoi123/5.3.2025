```python
#Question num 1
np.random.seed(80)  
matrix1 = np.random.randint(1, 100, (4, 4)) 

print(matrix1)


num = matrix1[matrix1 > 50]
print(num)

print('\nsum axis=0 columns:\n', matrix1.sum(axis=0))  

print('\nsum axis=1 rows:   \n', matrix1.sum(axis=1))

print("\nMax row:", np.max(matrix1, axis=1))
print("Max column:", np.max(matrix1, axis=0))


print("\nMin row:", np.min(matrix1, axis=1))
print("Min column:", np.min(matrix1, axis=0))

print("\nMean row:", np.mean(matrix1, axis=1))
print("Mean column:", np.mean(matrix1, axis=0))
```
